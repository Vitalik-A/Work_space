```python
ing=set()
pop_ing=[]
for i in recipes:
    for l in i['ingredients']:
        ing.add(l)
food={}
for item in ing:
    food[item]=0
for i in recipes:
    for item in i['ingredients']:
        food[item]+=1
for k,v in food.items():
    if v>100:
        pop_ing.append(k)
pop_ing        
```

```python
max=0
ingr=''
for k,v in food.items():
    if v>max:
        max=v
        ingr=k
print(ingr, max)        
```

```python
#12.4
with open('recipes.json') as f:
   recipes = json.load(f)
df = pd.DataFrame(recipes)

ing=set()
for i in recipes:
    for l in i['ingredients']:
        ing.add(l)

def find_item(cell):
    if item in cell:
        return 1
    return 0
for item in ing:
    df[item] = df['ingredients'].apply(find_item)
df['ingredients'] = df['ingredients'].apply(lambda x: len(x))   
df.head()
```

```python
#12.5
df.to_csv('recipes.csv', index = False)
```

```python
df=pd.read_csv('recipes.csv')
ids=df.id.unique()
ingredients=df.columns
ingredients=ingredients[3:]
new_recipes = []

def make_list(x):
    lis=[]
    for i in ingredients:
        if x[i].iloc[0]==1:
            lis.append(i)
    return lis
for current_id in ids:
    cuisine = df[df['id'] == current_id]['cuisine'].iloc[0]
    current_ingredients = make_list(df[df['id'] == current_id])
    current_recipe = {'cuisine': cuisine, 'id': int(current_id), 'ingredients': current_ingredients}
    new_recipes.append(current_recipe)
new_recipes    
```

```python
with open("new_recipes.json", "w") as write_file:
    json.dump(new_recipes, write_file)
```
