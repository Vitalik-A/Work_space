```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
myData = pd.read_csv('mycar.csv')
```

```python
myData.head(5)
```

```python
X = myData.iloc[:,:-1].values
Y = myData.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.3)
```

```python
y_pred = myModel.predict(X_test)
y_pred
```

```python
from sklearn import metrics  # подгружаем метрики
MSE = metrics.mean_squared_error(Y_test, y_pred)
MSE
```

```python
data = pd.read_csv('data_flats.csv', sep =';')
data.head()
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(20,12))
sns_heatmap = sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
data.price_doc.hist()
```

```python
data['price_doc'] = data['price_doc'].apply(lambda w: np.log(w + 1))
data.price_doc.hist();
```

```python
sns.set(font_scale=2)
plt.subplots(figsize=(20,20))
sns.heatmap(data.corr(),square=True, annot=True, 
            fmt=".1f", linewidths=0.1, cmap="RdBu")
```

```python
data.columns
```

```python
data.drop(columns=['id','life_sq','preschool_education_centers_raion','kindergarten_km',
                  'park_km','kremlin_km'],inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X=data.drop(columns=['price_doc'])
Y=data['price_doc']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2,random_state=77)
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
X_train.head()
```

```python
scalter=RobustScaler()
```

```python
X_train_new=scalter.fit_transform(X_train)
X_test_new=scalter.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
myModel = LinearRegression() #Обозначаем, что наша модель - линейная регрессия
myModel.fit(X_train_new,Y_train) #обучаем модель на обучающих данных
```

```python
y_pred = myModel.predict(X_test_new)
y_pred
```

```python
Y_test1=np.exp(Y_test)
y_pred1=np.exp(y_pred)
```

```python
from sklearn import metrics  # подгружаем метрики
MSE = metrics.mean_squared_error(Y_test1, y_pred1)
MSE
```

```python
from sklearn.datasets import load_boston
```

```python
#3.6.1
data = load_boston()
```

```python
X,Y=data['data'],data['target']
X = np.hstack([np.ones(X.shape[0])[:, np.newaxis], X])
```

```python
def linar_reg(x,y):
    thet=np.linalg.inv(x.T@x)@x.T@y
    return thet
```

```python
theta=linar_reg(X,Y)
y_pred=X@theta
```

```python
from sklearn import metrics  # подгружаем метрики
MSE = metrics.mean_squared_error(Y, y_pred)
RMSE=np.sqrt(MSE)
print(f'RMSE={RMSE}')
```

```python
#3.6.2
from sklearn.linear_model import LinearRegression
myModel = LinearRegression() #Обозначаем, что наша модель - линейная регрессия
myModel.fit(X,Y) #обучаем модель на обучающих данных
```

```python
y_pred = myModel.predict(X)
MSE = metrics.mean_squared_error(Y, y_pred)
RMSE=np.sqrt(MSE)
print(f'RMSE={RMSE}')
```
